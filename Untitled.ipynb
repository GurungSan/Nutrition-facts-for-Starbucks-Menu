### What is the average daily intake percentage of each nutrition type for Starbucks drink items?

```python
daily_intake = starbucks.copy(deep=True)
daily_intake.calories = daily_intake.calories / 2000 * 100
daily_intake.fat = daily_intake.fat / 78 * 100
daily_intake.cholesterol = daily_intake.cholesterol / 300 * 100
daily_intake.sodium = daily_intake.sodium / 2300 * 100
daily_intake.carb = daily_intake.carb / 275 * 100
daily_intake = daily_intake.drop(columns=['sugar']) # no daily level for total sugars
daily_intake.protein = daily_intake.protein / 50 * 100
daily_intake.caffeine = daily_intake.caffeine / 400 * 100
```

```python
daily_intake.head()
```

```python
#average daily intake
daily_intake.mean()
```

```python
# max intake 
daily_intake.max()
```

```python
# Highest daily intake % of caffeine
daily_intake.nlargest(10, 'caffeine')[['drink_name', 'caffeine']]
```

```python
# Highest daily intake % of calories
daily_intake.nlargest(10, 'calories')[['drink_name', 'calories']]
```

```python
# Highest daily intake % of carbs (includes sugar)
daily_intake.nlargest(10, 'carb')[['drink_name', 'carb']]
```

```python
heatmap = daily_intake.groupby('type').mean()

plt.figure(figsize=(8,6))
plt.title('Average Nutrition Daily Intake % of Starbucks Items by Category', fontsize=16)
sns.heatmap(heatmap, cmap="Greens", annot=True)
```
